```python

price_dist_plot = sns.histplot(df2['price'], bins=50, kde=True)
plt.title('Price Distribution')
plt.xlabel('price')
plt.ylabel('?')
#suggested by copilot
plt.xlim(0, 1_500_000)
plt.xticks(np.arange(0, 1_500_001, 250_000))
plt.grid(axis='y', linestyle='--', alpha=0.3)

plt.show(price_dist_plot)
```

```python
bath_plot = sns.histplot(df2['bathrooms'], bins=50, kde=True)
plt.title('Bathrooms Distribution')
plt.xlabel('number of bathrooms')
plt.show(bath_plot)
```

```python
room_plot = sns.histplot(df2['bedrooms'], bins=50, kde=True)
plt.title('Bedrooms Distribution')
plt.xlabel('number of bedrooms')
plt.xlim(0, 10)
plt.xticks(np.arange(0, 11, 1))
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.show(room_plot)
```

```python
living_area_plot = sns.histplot(df2['sqft_living'], bins=50, kde=True)
plt.title('Living Area Distribution')
plt.xlabel('square feet of living area')
plt.xlim(0, 10_000)
plt.xticks(np.arange(0, 10_001, 1_000))
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
#suggested by copilot
plt.subplots_adjust(top=0.9)  # Adjust the top margin to fit the title
plt.ylabel('Frequency')
plt.xlabel('Square Feet of Living Area')
plt.show(living_area_plot)
```

```python
sns.regplot(x='price', y='sqft_living', data=df2, line_kws={'color': 'red'})
plt.title('Regression Plot of Price vs. Sqft Living')
plt.xlabel('Price $')
plt.ylabel('Living Space')
plt.grid(True)
plt.show()
```

```python
sns.regplot(x='sqft_lot', y='price', data=df2, line_kws={'color': 'red'})
plt.title('Regression Plot of Price vs. Sqft Lot')
plt.xlabel('Price $')
plt.ylabel('Living Space')
plt.grid(True)
plt.show()
```

```python
categoric_columns = ['waterfront', 'view', 'condition', 'grade']
for column in categoric_columns:
    plt.figure(figsize=(5, 3))
    sns.countplot(x=column, data=df2, palette='viridis')
    plt.title(f'Count Plot of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.3)
    plt.tight_layout()
    plt.show()
```
